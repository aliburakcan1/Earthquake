```python
import re

def lower(self):
    self = re.sub(r"Ä°", "i", self)
    self = re.sub(r"I", "Ä±", self)
    self = re.sub(r"Ã‡", "Ã§", self)
    self = re.sub(r"Å", "ÅŸ", self)
    self = re.sub(r"Ãœ", "Ã¼", self)
    self = re.sub(r"Ä", "ÄŸ", self)
    self = self.lower() # for the rest use default lower
    return self

from sklearn.feature_extraction.text import CountVectorizer
c_vec = CountVectorizer(ngram_range=(1,1))
# matrix of ngrams
ngrams = c_vec.fit_transform(tweets_df['tweets'])
# count frequency of ngrams
count_values = ngrams.toarray().sum(axis=0)
# list of ngrams
vocab = c_vec.vocabulary_
df_ngram = pd.DataFrame(sorted([(count_values[i],k) for k,i in vocab.items()], reverse=True)
            ).rename(columns={0: 'frequency', 1:'bigram/trigram'})
df_ngram
```
