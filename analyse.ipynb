```python
import pandas as pd
import numpy as np

tweets_df = pd.read_excel("earthquake/labeled_tweets_excel.xlsx", usecols=['Datetime', 'Text', 'retweetCount', 'rescue']) #.sample(frac=.1)
tweets_df = tweets_df.drop_duplicates(subset=['Text'], keep='last')
tweets_df = tweets_df[
    (
    (tweets_df['rescue'] == 0) 
        & (tweets_df['Text'].str.contains('deprem', case=False)
          & (tweets_df['retweetCount'] > 100))
    ) | (tweets_df['rescue'] == 1)
]
print(tweets_df.rescue.value_counts())
tweets_df
```

```python
tweets_df['Text'] = tweets_df['Text'].str.replace('@AFADBaskanlik', ' afadmention ', regex=False)
tweets_df
```

```python
tweets_df = tweets_df.dropna(subset=['Text'])
tweets_df
```
